Disadvantages:

2. Selection Sort

Simplicity: Like Bubble Sort, Selection Sort is easy to implement and understand. It repeatedly selects the smallest (or largest) element and places it in the correct position.

In-place Sorting: Selection Sort works in-place, meaning it doesn't require additional memory for another list (space complexity
O(1)).

Disadvantages:

Small datasets where simplicity is preferred.
